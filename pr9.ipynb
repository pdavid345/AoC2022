```python
with open('input/9.test') as f:
    dat = [i.split(' ') for i in f.read().split('\n')]

# print(dat)
```

```python
# Update tail position to follow head
def updateTail(H,T):
    # X direction follow tail
    if(abs(H[0] - T[0]) == 2):

        T[0] = H[0] - 1 + 2*(T[0] > H[0])
        if (T[1] < H[1]):
            T[1] +=1
        elif (T[1] > H[1]):
            T[1] -= 1
    elif(abs(H[1] - T[1]) == 2):

        T[1] = H[1] - 1 + 2*(T[1] > H[1])
        if (T[0] < H[0]):
            T[0] +=1
        elif (T[0] > H[0]):
            T[0] -= 1
    return T
```

```python
H = [0,0]
T = [0,0]
tailLocation = set()
tailLocation.add(tuple(T))
for l in dat:
    if l[0] == 'R':
        for i in range(int(l[1])):
            H[0] = H[0] + 1
            T = updateTail(H,T)
            tailLocation.add(tuple(T))
            
    elif l[0] == 'L':
        for i in range(int(l[1])):
            H[0] = H[0] - 1
            T = updateTail(H,T)
            tailLocation.add(tuple(T))
    elif l[0] == 'U':
        for i in range(int(l[1])):
            H[1] = H[1] -1
            T = updateTail(H,T)
            tailLocation.add(tuple(T))
    elif l[0] == 'D':
        for i in range(int(l[1])):
            H[1] = H[1] + 1
            T = updateTail(H,T)
            tailLocation.add(tuple(T))
    # print(H, T)
print(f'Part 1: {len(tailLocation)}')
```

```python
def plotSnake(body):
    xdats = [x[0] for x in body]
    ydats = [x[1] for x in body]
    xdats = [i - min(xdats) for i in xdats]
    ydats = [i - min(ydats) for i in ydats]
    # print(xdats)
    # print(ydats)

    plotx = max(xdats) + 2
    ploty = max(ydats) + 2

    out = [['.' for i in range(ploty)] for j in range(plotx)]

    for i in range(len(xdats)):
        out[xdats[i]][ydats[i]] = i
    
    for i in range(plotx):
        print(' '.join([str(j) for j in out[i]]))




    

```
